## SVM Model (Attempt 3 Models)

```python
print("Importing Libraries...")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pickle
import time
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import math
from sklearn.metrics import r2_score
print("Successfully imported!")
```

```python
print("Reading Datafile...")
checkpoint5 = pd.read_csv('checkpoint-5.csv', nrows=100000)
print("Success!")
```

```python
checkpoint5
```

```python
model1 = checkpoint5.sample(frac = 0.267)
model2 = checkpoint5.drop(model1.index).sample(frac = 0.267)
model3 = checkpoint5.drop(model1.index).drop(model2.index).sample(frac = 0.267)
overall_test = checkpoint5.drop(model1.index).drop(model2.index).drop(model3.index)
```

```python
# model1.reset_index()
# model2.reset_index()
# model3.reset_index()
```

```python
print("Obtaining label for 1...")
cols = 'fare_amount'
label1 = model1[[cols]]
label2 = model2[[cols]]
label3 = model3[[cols]]
overalltest_label = overall_test[[cols]]
print("Obtained labels!")
```

```python
# del checkpoint5['dist']
del model1['time']
del model1['fare_amount']
del model1['pickup_longitude']
del model1['pickup_latitude']
del model1['dropoff_longitude']
del model1['dropoff_latitude']

del model2['time']
del model2['fare_amount']
del model2['pickup_longitude']
del model2['pickup_latitude']
del model2['dropoff_longitude']
del model2['dropoff_latitude']

del model3['time']
del model3['fare_amount']
del model3['pickup_longitude']
del model3['pickup_latitude']
del model3['dropoff_longitude']
del model3['dropoff_latitude']

del overall_test['time']
del overall_test['fare_amount']
del overall_test['pickup_longitude']
del overall_test['pickup_latitude']
del overall_test['dropoff_longitude']
del overall_test['dropoff_latitude']
```

### Normalising the values

```python
X_sc1 = StandardScaler()
y_sc1 = StandardScaler()

model1 = X_sc1.fit_transform(model1)
label1 = y_sc1.fit_transform(label1)

print('Fitting the model1...')
regressor1 = SVR(kernel = 'rbf')
regressor1.fit(model1, label1.ravel())
print('Succeeded in fitting model1!')
#--------------------------------------------------

X_sc2 = StandardScaler()
y_sc2 = StandardScaler()

model2 = X_sc2.fit_transform(model2)
label2 = y_sc2.fit_transform(label2)

print('Fitting the model2...')
regressor2 = SVR(kernel = 'rbf')
regressor2.fit(model2, label2.ravel())
print('Succeeded in fitting model2!')
#--------------------------------------------------


X_sc3 = StandardScaler()
y_sc3 = StandardScaler()

model3 = X_sc3.fit_transform(model3)
label3 = y_sc3.fit_transform(label3)

print('Fitting the model2...')
regressor3 = SVR(kernel = 'rbf')
regressor3.fit(model3, label3.ravel())
print('Succeeded in fitting model2!')

#--------------------------------------------------
```

```python
print('Predicting...for 1 ')
y_pred1 = regressor1.predict(X_sc1.transform(overall_test))
y_pred1 = y_sc1.inverse_transform(y_pred1.reshape(-1, 1))
print('Predicted all!')

print('Predicting...for 2 ')
y_pred2 = regressor2.predict(X_sc2.transform(overall_test))
y_pred2 = y_sc2.inverse_transform(y_pred2.reshape(-1, 1))
print('Predicted all!')

print('Predicting...for 3 ')
y_pred3 = regressor3.predict(X_sc3.transform(overall_test))
y_pred3 = y_sc3.inverse_transform(y_pred3.reshape(-1, 1))
print('Predicted all!')
```

```python
overalltest_label = overalltest_label.to_numpy()
overalltest_label = overalltest_label.flatten()
y_pred1 = y_pred1.flatten()
y_pred2 = y_pred2.flatten()
y_pred3 = y_pred3.flatten()
```

```python
print("Length of prediction is:", len(y_pred1))
combined_pred = []
for x in range(len(y_pred1)):
    average = (float("{0:.2f}".format(y_pred1[x])) +  float("{0:.2f}".format(y_pred2[x])) +  float("{0:.2f}".format(y_pred3[x]))) / 3
    combined_pred.append(average)
```

```python
df = pd.DataFrame({'Predicted value1': y_pred1,'Predicted value2': y_pred2,'Predicted value3': y_pred3, 'Average': combined_pred,'Real Value': overalltest_label})
df
df.to_csv('WithDist_SVM/Dist3Models_predictions_SVR200000.csv',index=False )
```

```python
#Get Report
print("Score for classifier:", r2_score(combined_pred,overalltest_label))
# print ("Time taken to train:", time_train)
# print("Time taken to predict",time_predict)
print("Mean square error is -> ",mean_squared_error(combined_pred,overalltest_label))
print("Root Mean square error is -> ",math.sqrt(mean_squared_error(combined_pred,overalltest_label)))
print("Mean absolute error is -> ",mean_absolute_error(combined_pred,overalltest_label))

# Save report 
report = { 
    'version': 'SVM with Dist',
    'number_rows': 200000,
    'r2_score': r2_score(combined_pred,overalltest_label),
    'MSE': mean_squared_error(combined_pred,overalltest_label),
    'RMSE': math.sqrt(mean_squared_error(combined_pred,overalltest_label)),
    'MAE': mean_absolute_error(combined_pred,overalltest_label)}

final_report = pd.DataFrame.from_dict(report.items())
final_report.to_csv('WithDist_SVM/Dist3Models_report_SVR200000.csv',index=False )
```
