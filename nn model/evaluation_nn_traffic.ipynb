```python
import matplotlib.pyplot as plt
import tensorflow as tf
import pandas as pd
import ast
from tensorflow.keras.applications.resnet import ResNet50
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
model = tf.keras.models.load_model('NN_traffic_density.h5')
model
```

```python
checkpoint5 = pd.read_csv('../checkpoint5_traffic.csv')
checkpoint5
```

```python
del checkpoint5['fare_amount']
del checkpoint5['time']
```

```python
checkpoint5
```

```python
print("scaling...")
#FEATURE SCALING - to scale the data to ensure a suitable range compatible with the model
sc = StandardScaler()
checkpoint_5_scaled = np.array(sc.fit_transform(checkpoint5))
print("successfully scaled...")
```

```python
checkpoint_5_scaled
```

```python
print("Reading Datafile...")
test_set = pd.read_csv('../kaggle_test_traffic.csv')
print("Success!")
```

```python
test_set
```

```python
test_key_series = test_set['key'].squeeze()
del test_set['key']
```

```python
test_set
```

```python
test_set_scaled = np.array(sc.transform(test_set))
```

```python
# make a prediction
y_pred = model.predict_on_batch(test_set_scaled)
# show the inputs and predicted outputs
```

```python
y_pred
```

```python
y_pred = y_pred.flatten()
```

```python
new_y_pred= []
for a in y_pred: 
    new_value = float("%.2f" % a)
    new_y_pred.append(new_value)
y_pred_series = pd.Series(new_y_pred)
df = pd.DataFrame({'key': test_key_series,'fare_amount': new_y_pred})
df
```

```python
df.to_csv('NNPrediction_traffic.csv',index=False )
```
