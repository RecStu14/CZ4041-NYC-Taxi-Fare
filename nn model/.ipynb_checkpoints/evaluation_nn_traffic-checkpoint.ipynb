```python
print("scaling...")
#FEATURE SCALING - to scale the data to ensure a suitable range compatible with the model
sc = StandardScaler()
checkpoint_5_scaled = np.array(sc.fit_transform(checkpoint5))
print("successfully scaled...")
```

```python
print("Reading Datafile...")
test_set = pd.read_csv('../processed_test.csv')
print("Success!")
```

```python
test_set
```

```python
y_pred = y_pred.flatten()
```

```python
new_y_pred= []
for a in y_pred: 
    new_value = float("%.2f" % a)
    new_y_pred.append(new_value)
y_pred_series = pd.Series(new_y_pred)
df = pd.DataFrame({'key': test_key_series,'fare_amount': new_y_pred})
df
```

```python
#reading the history dictionary
# reading the data from the file
with open('history_s1.txt') as f:
    data = f.read()
```

```python
s1_history = ast.literal_eval(data)
print("Type:", type(s1_history))
s1_history
```

```python
s1_history.keys()
```

```python
training_loss = s1_history['loss']
val_loss = s1_history['val_loss']

training_accuracy = s1_history['categorical_accuracy']
val_accuracy = s1_history['val_categorical_accuracy']

epochs = range(1,201)
```

```python
#Plotting the Loss Curves
plt.plot(epochs, training_loss, 'g', label='Training Loss')
plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.title('Training Phase - Loss Curves')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
#Plotting the Loss Curves
plt.plot(epochs, training_accuracy, 'g', label='Training Accuracy')
plt.plot(epochs, val_accuracy, 'b', label='Validation Accuracy')
plt.title('Training Phase - Accuracy Curves')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
